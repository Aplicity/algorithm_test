在一个$n*n$的棋盘中标有一条没有分叉的路线，一个骰子被放置在起点。骰子的六个面涂有六种不同的颜色，分别计为1-6。骰子的底面大小和棋盘的格子大小一致，骰子沿路线进行不打滑的滚动，骰子的底面和棋盘接触时会在棋盘上留下对应的颜色。请你输入骰子滚到终点后路径中每个位置的颜色。

### 解决

```python
class Dice(object):
    def __init__(self, upper,lower,left,right,forward,backward):
        self.upper =  upper
        self.lower = lower
        self.left  = left
        self.right = right
        self.forward = forward
        self.backward = backward
        
    def forward_roll(self):
        temp_upper = self.upper
        temp_lower = self.lower
        temp_left = self.left
        temp_right = self.right
        temp_forward = self.forward
        temp_backward = self.backward
        
        self.upper = temp_forward
        self.forward = temp_lower
        self.lower = temp_backward
        self.backward = temp_upper
        
    def backward_roll(self):
        temp_upper = self.upper
        temp_lower = self.lower
        temp_left = self.left
        temp_right = self.right
        temp_forward = self.forward
        temp_backward = self.backward
        
        self.upper = temp_backward
        self.forward = temp_upper
        self.lower  = temp_forward
        self.backward = temp_lower
        
    def left_roll(self):
        temp_upper = self.upper
        temp_lower = self.lower
        temp_left = self.left
        temp_right = self.right
        temp_forward = self.forward
        temp_backward = self.backward
        
        self.upper = temp_right
        self.right  = temp_lower
        self.lower = temp_left
        self.left =  temp_upper
        
    def right_roll(self):
        temp_upper = self.upper
        temp_lower = self.lower
        temp_left = self.left
        temp_right = self.right
        temp_forward = self.forward
        temp_backward = self.backward     
        
        self.upper = temp_left
        self.right = temp_upper
        self.lower = temp_right
        self.left  = temp_lower
        
    def roll(self,direction):
        if direction == 'right':
            self.right_roll()
        if direction == 'left':
            self.left_roll()
        if direction == 'forward':
            self.forward_roll()
        if direction == 'backward':
            self.backward_roll()

def get_operate(row, col, chessboard, passed_XYs):
    '''
    获取骰子下一步应该翻滚的方向和翻滚后的坐标位置
    row : 当前骰子所在棋盘的行索引
    col : 当前骰子所在棋盘的列索引
    chessboard : 棋盘轨迹数据
    passed_XYs : 骰子已走过的轨迹坐标list
    '''
    n = len(chessboard)
    candidate_xy = []
    forward_xy = (row - 1, col)
    backward_xy = (row + 1, col)
    left_xy = (row, col - 1)
    right_xy = (row, col + 1)
    for xy in [forward_xy, backward_xy, left_xy, right_xy]:
        x, y = xy
        if x >= 0 and x < n and y >= 0 and y < n:  # 判断是否在棋盘范围内
            if (chessboard[x][y] == '#') or (chessboard[x][y] == 'E') : 
                if xy not in passed_XYs: # 下一个位置在没走过的轨迹上
                    if xy == forward_xy:
                        return 'forward', x, y
                    if xy == backward_xy:
                        return 'backward', x, y
                    if xy == left_xy:
                        return 'left', x, y
                    if xy == right_xy:
                        return 'right', x, y


def find_start_xy(chessboard):
    '''寻找起始点坐标位置'''
    n = len(chessboard)
    for i in range(n):
        for j in range(n):
            if chessboard[i][j] == 'S':
                return i,j
                break
```

## Q2
### 问题描述
最近小强很喜欢玩魔塔，勇者和魔塔的怪物都有三个属性：血量、攻击力和防御力。当勇者遭遇一个怪物时，战斗方式如下：
* 由勇者先手攻击，回合制，轮流对对方造成伤害。
* 伤害值为max(伤害方的攻击力-对方的防御力,1)。其中max(a,b)代表a和b中较大的值。
* 当一方的血量降低到0以下时，战斗立即结束，另一方胜利。

现在魔塔中的怪物排成一排，勇者需要按照顺序进行战斗，当勇者成果击败一只怪物后他可以获得奖励：从以下三项中选择一项。
* 1. 血量增加1000
* 2. 攻击力增加10
* 3. 防御力增加10

现在小强想知道，在最优策略下勇者最多能击败多少只怪物。

### 输入描述
第一行4个空格分隔的正整数 $H,A,D,n$ 分别代表勇者的血量，攻击力，防御力，怪物的数量。

接下来n行每行三个空格分隔的正整数 $H_i,A_i,D_i$ 用来描述第$i$只怪物的血量，攻击力，防御力。

$1 \leq n \leq 200$
$1 \leq H,A,D,H_i,A_i,D_i \leq 10^9$

### 输出描述
仅一行一个整数表示最大击败怪物数。

### 示例
#### 输入

10 10 10 3

6 11 9

2 10000 10

100 10000 10000

#### 输出
2

### 示例说明
击败第一只怪物勇者损耗5血量，接下里最优策略是选择提高10点攻击力，这样在对战第二只怪物时一次攻击击败第二只怪物，损耗血量为0，接下来无论选择说明奖励均无法战胜第三只怪物，所以最多击败怪物数为2。

```python
test_file = '10 10 10 3\n6 11 9\n2 10000 10\n100 10000 10000'
print(test_file)
```

### 解决

```python
class Character(object):
    def __init__(self, H,A,D):
        self.H = H
        self.A = A
        self.D = D
        
    def choice_tactic(self,k):
        if k == 1:
            self.H += 1000
        if k == 2:
            self.A += 10
        if k == 3:
            self.D += 10 
            
    def be_attack_by(self,obj):
        self.H -= max(obj.A - self.D , 1)
        
    
    def update_params_after_battle(self, obj):
        while self.H > 0 and obj.H > 0 :
            obj.H -= max(self.A - obj.D, 1)
            if obj.H <= 0 :
                break
            else:
                self.be_attack_by(obj)
        H,A,D = self.H, self.A, self.D
        return H,A,D
    
def get_all_plan(n):
    '''穷尽出所有方案'''
    ret = [[]]
    for i in range(n):
        temp = []
        for item in ret:
            temp.append(item + [1])
            temp.append(item + [2])
            temp.append(item + [3])
        ret = temp
    return ret 
```

```python
test_file = '10 10 10 3\n6 11 9\n2 10000 10\n100 10000 10000'

items = []
for i,line in enumerate(test_file.split('\n')):
    lineArr = line.split()
    
    if i == 0:
        my_H, my_A, my_D ,n = lineArr
        my_H, my_A, my_D, n = int(my_H), int(my_A), int(my_D), int(n)
    if i  > 0:
        H,A,D = lineArr
        H,A,D = int(H), int(A), int(D)
        items.append([H,A,D])

max_deep = 0
n = 5
plans = [[]]
for i in range(n):
    temp = []
    for item in plans :
        temp.append(item + [1])
        temp.append(item + [2])
        temp.append(item + [3])
    plans = temp
    for plan in plans:
        Me = Character(my_H,my_A,my_D)
        local_deep = 0
        for k in range(i+1):
            h,a,d = items[k]
            Monster = Character(h,a,d)
            Me.update_params_after_battle(Monster)
            if Me.H > 0 :
                local_deep += 1 
                if local_deep > max_deep:
                    max_deep = local_deep
                    best_plan = plan[:max_deep-1]
                tactic = plan[k]
                Me.choice_tactic(tactic)
            else:
                plans.remove(plan)
                break
    
print(max_deep)
    
```
